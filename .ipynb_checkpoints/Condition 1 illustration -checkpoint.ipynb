### Condition 1

Condition 1 has this form:
$$\forall k,\; c_n(k) = \frac{1}{n} \max_{A\in\mathcal{A}_{k}(n)}\max_{B\in\mathcal{Z}_A} \frac{p(A)}{p(B)} <\infty $$

**Pitman--Yor process** $$\frac{1}{n} \frac{p(A)}{p(B)} =\frac{k+1}{n(\theta+k\sigma)} (1-\sigma+|A_\ell|-2)$$
So $$c_n(k) = (k+1)\frac{n-k}{n(\theta+k\sigma)}$$

**Dirichlet process**

As the Diichlet process is the particular case of Pitman-Yor process with $\sigma=0$,
$$c_n(k) = (k+1)\frac{n-k}{n\alpha}$$

**Gibbs-type process**

 \begin{align*}
        \frac{1}{n} \frac{p(A)}{p(B)} & = \frac{1}{n} \frac{V_{n,k}}{k!} \prod_{i=1}^k (1-\sigma)_{|A_i|-1} \frac{(k+1)!}{V_{n,k+1}} \left(\prod_{i=1}^{k+1} (1-\sigma)_{|B_i|-1}\right)^{-1} \\
        & =\frac{k+1}{n}\frac{V_{n,k}}{V_{n,k+1}} (1-\sigma+|A_\ell|-2)
    \end{align*}
So $$c_n(k) = \frac{\sigma+n-k}{n(k+1)}\times\frac{V_{n,k}}{V_{n,k+1}}$$


**Dirichlet multinomial process**
$$ \frac{1}{n}\frac{p(A)}{p(B)} = \frac{c + K(|A_\ell| -1)  }{nc(K-k)} $$
So $$c_n(k) = \frac{c+K(n-k)}{nc(K-k)}$$

Illustrations of Condition 1, for $k\in\{1,10,100\}$: we plot the function $n\mapsto c_n(k)$ for the different processes and $n\in \{1,\ldots,5000\}$.

```julia
k = 1
n_range = collect(range(1,1000, step = 10))
# Store the functions cn(1) for the different processes
DF_1 = DataFrame(n = Int64[], proc = String[], ratio = Float64[])

for i in n_range
    n_l = i-k+1
    ratio_py = (σ-1+n_l)*(k+1)/(i*(θ+k*σ))
    push!(DF_1, [i, "PY", ratio_py])
    ratio_dpm = (c+K*(n_l-1))/(i*c*(K-k))
    push!(DF_1, [i, "DMP", ratio_dpm])
    ratio_G = (σ-1+n_l)/i*(k+1)*GibbsTypePriors.Vnk_NGG(i, k, β, σ)/GibbsTypePriors.Vnk_NGG(i, k+1, β, σ)
    ratio_G = convert(Float64, ratio_G)
    push!(DF_1, [i, "NGG", ratio_G])
    ratio_dp = (k+1)*(n_l-1)/(i*α)
    push!(DF_1, [i, "DP", ratio_dp])
end
```

```julia
R"proc = factor($(DF_1.proc), levels = c('DMP', 'NGG', 'PY', 'DP'))
p1 = ggplot(data.frame(x =$(DF_1.n),
                         ratio = $(DF_1.ratio),
                         proc= proc),
     aes(x=x, y = ratio, colour = proc)) +
 geom_line(size = 1) + xlab(TeX(sprintf('$n$'))) + ylab(TeX(sprintf('$c_n(k)$'))) +
 theme_minimal() + 
 theme (axis.title= element_text(size=rel(1.5)), 
        axis.text.x = element_text(size=rel(1.2)),
        axis.text.y = element_text(size=rel(1.2)), legend.position = 'none') +
 scale_color_viridis(discrete= 'TRUE', begin = 0.95, end = 0,option = 'D', name = 'Priors')
"
R"
# Exportation
ggsave('Illustration_1_cond1.pdf', 
       p1, 
       width = 15, 
       height = 10, 
       units = 'cm',
       dpi=500) "
R"p1"
```

#### Condition 1 for n-dependent concentration parameters

We choose the concentration parameter $\alpha_n$ such that $E[K_n]=25$ for all $n$ for a **Dirichlet process**.

```julia
n_range = collect(range(50, 2000, step = 20))
# Store the different values for 𝛼n depending on n
𝛼_n = Array{Float64, 2}(undef, 1, length(n_range))

for i in range(1, length(n_range))
    n = n_range[i]
    f(x) = convert(Float64, GibbsTypePriors.E_Dirichlet(n,x))-25
    𝛼_n[i] = find_zero(f,10)
end
#println(𝛼_n)
```

```julia
# Store the function cn(k) for 3 values of k and depending on n
k_range = [1,10,100]
DF_𝛼 = DataFrame(n = Int64[], ratio = Float64[], k = Int64[])

for i in range(1, length(n_range))
    n = n_range[i]
    for k in k_range
        if k<n
            n_l=n-k+1
            ratio_dp = (k+1)*(n_l-1)/(i*𝛼_n[i])
            push!(DF_𝛼, [n, ratio_dp, k])
        end
    end
end
```

```julia
R"p3 = ggplot(data.frame(x =$(DF_𝛼.n),
                         ratio = $(DF_𝛼.ratio),
                         k = factor($(DF_𝛼.k))),
     aes(x=x, y = ratio, colour = k)) +
 geom_line(size = 1) + xlab(TeX(sprintf('$n$'))) + ylab(TeX(sprintf('$c_n(k)$')))+
 ggtitle(TeX(sprintf('Value of $c_n(k)$ for DP with $\\alpha_n$ varying')))+
 theme_minimal() + 
 theme (axis.title= element_text(size=rel(1.5)), legend.title = element_text(size=rel(1.5)),
        axis.text.x = element_text(size=rel(1.2)),
        axis.text.y = element_text(size=rel(1.2)), legend.text = element_text(size=rel(1.2))) +
 scale_color_viridis(discrete= 'TRUE', begin = 0.95, end = 0, option = 'D', name = 'k') +
 theme(plot.title = element_text(hjust = 0.5, size=rel(0.8)))
"
```

We do the same, choosing the concentration parameter $c_n$ such that $\mathbb{E}[K_n]=25$ for all $n$ for a **Dirichlet multinomial process**.

```julia
K = 200
# Store the different values for cn depending on n
c_n = Array{Float64, 2}(undef, 1, length(n_range))

for i in range(1, length(n_range))
    n = n_range[i]
    f(x) = convert(Float64, GibbsTypePriors.E_Dirichlet_multinomial(n,K,x))-25
    c_n[i] = find_zero(f, 10)
end
#println(c_n)
```

```julia
# Store the function cn(k) for 3 values of k and depending on n
DF_c = DataFrame(n = Int64[], ratio = Float64[], k = Int64[])

for i in range(1, length(n_range))
    n = n_range[i]
    for k in k_range
        if k<n
            n_l=n-k+1
            ratio_dmp = (c_n[i]+K*(n_l-1))/(i*c_n[i]*(K-k))
            push!(DF_c, [n, ratio_dmp, k])
        end
    end
end
```

```julia
R"p3 = ggplot(data.frame(x =$(DF_c.n),
                         ratio = $(DF_c.ratio),
                         k = factor($(DF_c.k))),
 aes(x=x, y = ratio, colour = k)) +
 geom_line(size = 1) + xlab(TeX(sprintf('$n$'))) + ylab(TeX(sprintf('$c_n(k)$'))) +
 ggtitle(TeX(sprintf('Value of $c_n(k)$ for DMP with $c_n$ varying')))+
 theme_minimal() + 
 theme (axis.title= element_text(size=rel(1.5)), legend.title = element_text(size=rel(1.5)),
        axis.text.x = element_text(size=rel(1.2)),
        axis.text.y = element_text(size=rel(1.2)), legend.text = element_text(size=rel(1.2))) +
 scale_color_viridis(discrete= 'TRUE', begin = 0.95, end = 0, option = 'D', name = 'k') +
 theme(plot.title = element_text(hjust = 0.5, size=rel(0.8)))
"
```
